```python
import datetime

def send_report():
    month = datetime.datetime.now().month
    if month % 2 == 0:
        print("Отчет отправлен!")
    else:
        print("Ошибка: отчеты отправляются только в четные месяцы.")

def backup_data():
    month = datetime.datetime.now().month
    if month % 2 == 0:
        print("Бэкап создан!")
    else:
        print("Ошибка: бэкап доступен только в четные месяцы.")
```

```python
import datetime

def only_even_months(func):
    def wrapper(*args, **kwargs):
        current_month = datetime.datetime.now().month
        if current_month % 2 == 0:
            return func(*args, **kwargs)
        else:
            print(f"Отказ: Функция '{func.__name__}' не запущена (сейчас {current_month}-й месяц, а нужен четный).")
    return wrapper

@only_even_months
def send_report():
    print("Выполняю: Отчет успешно отправлен.")

@only_even_months
def backup_data():
    print("Выполняю: Данные скопированы в архив.")

send_report()
backup_data()
```

```python
import sys

def linux_only(func):
    def wrapper(*args, **kwargs):
        # sys.platform возвращает 'linux', 'win32' или 'darwin' (macOS)
        if sys.platform == 'linux':
            return func(*args, **kwargs)
        else:
            print(f"Ошибка: Функция '{func.__name__}' работает только на Linux. Ваша система: {sys.platform}")
    return wrapper

@linux_only
def install_package():
    print("Пакет успешно установлен в /usr/bin/")

@linux_only
def clear_cache():
    print("Кэш очищен.")

# Проверка:
install_package()
```

```python
@only_even_months
@linux_only
def critical_update():
    print("Внимание! Выполняется критическое обновление системы...")

# Эта функция запустится ТОЛЬКО если сейчас четный месяц И система — Linux.
critical_update()
```
